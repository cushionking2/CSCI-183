```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import io
import math
from sklearn.linear_model import LinearRegression

df2 = pd.read_csv("kc_house_data.csv")

df2
```

```python
#3. Find how many numerical features exist in the dataset.
#price,	bedrooms,	bathrooms,	sqft_living,	sqft_lot,	floors,	sqft_above,	sqft_basement,	yr_built,	yr_renovated,	sqft_living15,	sqft_lot15
df2n = df2.drop(['id', 'date', 'waterfront', 'view', 'condition', 'grade', 'zipcode', 'lat', 'long'], axis=1)
df2n
```

```python
df2.corr()
```

```python
#Correlation plot with all values
plt.subplots(figsize=(40,15))
sns.heatmap(
    df2.corr(),
    vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(20, 220, n=200),
    annot=True, annot_kws={'size': 15},
    square=True)
```

```python
plt.figure(figsize=(15, 5))
plt.plot(df2['bedrooms'], df2['price'], 'o') 
```

```python
predictors = ['bedrooms']
X = df2[predictors]
y = df2['price']

lm = LinearRegression()
model = lm.fit(X, y)

print(f'intercept = {model.intercept_}')
print(f'coefficient = {model.coef_}')

pred_y = model.predict(X)
print(pred_y.shape)
print(pred_y)
```

```python
# Plot regression against actual data
plt.figure(figsize=(15, 5))
plt.plot(df2['bedrooms'], df2['price'], 'o')
plt.plot(df2['bedrooms'], pred_y, 'r', linewidth=2)
plt.xlabel('bedrooms')
plt.ylabel('price')
plt.title('bedrooms vs price')

plt.show()
```

```python
sum = 0
data = df2[["bedrooms","price"]].values
for i,j in data:
    pred_value = 129802.35631825868 + i * 121716.12651184
    e = (pred_value - j) ** 2
    sum += e
print("total error of bedroons predictor:", sum) #total error
```
