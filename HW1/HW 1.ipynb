```python
#Aditya Ranade, Austin Wang CSCI 183
# HW1 - Problem 1 - Python Basics

def sum_of_primes(n):
    # initialize the sum
    sum = 0
    #initialize the count
    count = 0
    # list of the prime numbers, everything starts as true
    primes = [True] * (n+1)
    
    i = 2
    # loop through list and check to see if number is prime or not
    while i*i <= n:
        if primes[i] == True:
            # all rest of the multiples will be false
            j = 2 * i 
            while j <= n:
                # setting the multiples of the true prime number to be false
                primes[j] = False 
                j = j + i # increment
        i = i + 1 # increment
    
    # calculating the sum by doing the for loop in range of
    for i in range(2, n + 1):
        if primes[i]:
            count = count + 1
            sum = sum + i
    print("Number of primes: ",count)
    print("Sum of primes: ", sum)

# print the number and the sum of the prime numbers, example
n = 10
sum_of_primes(n)
```

```python
# HW1 - Problem 2 - Python Basics

def max_product(ls):
    # using brute force method of calculating each product, comparing, and then
    # checking
    
    # initialize the product
    max_prod = 0
    # loop through the first index till the last index of the given list
    for i in range(0, len(ls)-1):
        # if product of adjacent values is greater than current max product
        if (ls[i] * ls[i+1]) > max_prod:
            # if yes, then replace it with the larger product
            max_prod = ls[i] * ls[i+1]
    # return max product
    return max_prod

# return the max product of adjacent integers in the list, example
ls = [5,6,1,5,7,8,9,1,2,3,5,7,8,2,9]
print(max_product(ls))
    
```

```python
# HW1 - Problem 3 - Python Basics

import re

def is_valid(p):
    if not re.search("[A-Z]",p):
        return "Password does not contain a capital letter"
    elif not re.search("[a-z]",p):
        return "Password does not contain a lowercase letter"
    elif not re.search("[\^\$\*+?{}\[\]\|()!@#%]",p):
        return "Password does not contain a symbol"
    elif (len(p)<6 or len(p)>20):
        return "Password is not between 6 and 20 characters"
    else:
        return "Password is Valid!"

print(is_valid("sasssaa"))
print(is_valid("SASSSAA"))
print(is_valid("Sasssaa"))
print(is_valid("Sass!"))
print(is_valid("Sasssaa!"))
```

```python
# HW1 - Problem 4 - Python Basics

import re

def is_strong(p):
    if (len(p)<9 or len(p)>20):
        return "Password is not between 9 and 20 characters"
    elif re.search('[\s+]', p) or re.search('[\n+]', p): 
        # don't really need the + because one instance of space of newline is
        # enough to make the password not strong
        return "Password cannot contain a newline or space"
    elif re.search('(.)\\1{2,}', p):
        return "Password cannot contain the same character three times in a row"
    elif re.search('(..)(.*)\\1', p):
        return "Password cannot contain a repeated pattern"
    else:
        return "Password is Strong!"

print(is_strong("Zingerr1"))
print(is_strong("Zingerr1 2345"))
print(is_strong("Zin\ngerr12345"))
print(is_strong("Zingerrr12345"))
print(is_strong("ZingerZi12345"))
print(is_strong("Zingerr12345"))
```

```python
# HW1 - Problem 6 - Python for Data Science


import re

def find_words_with_length(n):
    ls = []
    with open('Data Science.txt','r') as f:
        ls = f.read().split()
    for i in range(0, len(ls)):
        ls[i] = re.sub(r'[^\w\s]','',ls[i])
    ls = [x for x in ls if len(x) == n]
    ls = set(ls)
    ls = list(ls)
    return ls

print(find_words_with_length(3))
print(find_words_with_length(4))
print(find_words_with_length(10))
```

```python
# HW1 - Problem 1 - Regular Expressions

import re

def find_z(s):
    if re.match('[^z]\w*z\w*[^z]', s):
        return "found z!"
    return "z not found!"
    
    
print(find_z("azzza"))
print(find_z("zaaaa"))
print(find_z("aaaaz"))
print(find_z("abczfgcaz"))
```

```python
# HW1 - Problem 2 - Regular Expressions

def find_csci(s):
    return re.findall('(CSCI\s\d+)', s)
    
    
print(find_csci("CSCI 10 MATH 9 MATH 11 CSCI 60 CSCI 182"))
print(find_csci("CSCI AB MATH 9 MATH 11 CSC 60 CSCI 182"))
```
